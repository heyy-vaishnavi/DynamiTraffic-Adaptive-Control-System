```python
import random as r
import matplotlib.pyplot as plt
import networkx as nx
import seaborn as sns
import numpy as np  
```

```python
class junction:
    def __init__(self, lil=[],ril=[],uil=[],dil=[],ll=None, rl=None, ul=None ,dl=None):
        self.lil=lil
        self.ril=ril
        self.uil=uil
        self.dil=dil
        self.ll=ll
        self.rl=rl
        self.ul=ul
        self.dl=dl
```

```python
j1=junction()
j2=junction()
j3=junction()
j4=junction()
```

```python
tl=[[[1,2,3],[2,3,4],[1,1,1],[2,2,2]],[[1,2,3],[2,3,4],[1,1,1],[2,2,2]],[[1,2,3],[2,3,4],[1,1,1],[2,2,2]],[[1,2,3],[2,3,4],[1,1,1],[2,2,2]]]
```

```python
def inp(j,i):
    j.lil=tl[i][0]
    j.uil=tl[i][1]
    j.ril=tl[i][2]
    j.dil=tl[i][3]
    #j.lil=list(eval(input("enter your vehicles")))
    #j.ril=list(eval(input("enter your vehicles")))
    #j.uil=list(eval(input("enter your vehicles")))
    #j.dil=list(eval(input("enter your vehicles")))
    return j
        
```

```python
jl=[j1,j2,j3,j4]
for i in range(4):
    jl[i]=inp(jl[i],i)
```

```python
j1.rl=j2
j1.dl=j3
j2.ll=j1
j2.dl=j4
j3.ul=j1
j3.rl=j4
j4.ll=j3
j4.ul=j2
```

```python
def priot(j):
    r=j.ril[1]*6 + j.ril[1]*2 + j.ril[0]
    l=j.lil[1]*6 + j.lil[1]*2 + j.lil[0]
    u=j.uil[1]*6 + j.uil[1]*2 + j.uil[0]
    d=j.dil[1]*6 + j.dil[1]*2 + j.dil[0]
    l1=[r,l,u,d]
    p1=['r','l','u','d']
    for i in range(4):
        for j in range(i,4):
            if(l1[i]<l1[j]):
                l1[i],l1[j]=l1[j],l1[i]
                p1[i],p1[j]=p1[j],p1[i]
    return p1
```

```python
def prioj(j):
    r=j.ril[1]*6 + j.ril[1]*2 + j.ril[0]
    l=j.lil[1]*6 + j.lil[1]*2 + j.lil[0]
    u=j.uil[1]*6 + j.uil[1]*2 + j.uil[0]
    d=j.dil[1]*6 + j.dil[1]*2 + j.dil[0]
    l1=[r,l,u,d]
    return sum(l1) 
```

```python
def tc(j):
    dic={'r':j.ril,'l':j.lil,'u':j.uil,'d':j.dil}    
    pl=priot(j)
    print("priority",pl)
    for i in pl:
        lq=len(dic[i])
        for o in range (lq):
            for k in range(dic[i][o]):
                di=r.randint(0,4)
                if(di==0):
                    if(j.ul!=None):
                        j.ul.dil[o]+=1
                if(di==1):
                    if(j.rl!=None):
                        j.rl.lil[o]+=1
                if(di==2):
                    if(j.dl!=None):
                        j.dl.uil[o]+=1
                if(di==3):
                    if(j.ll!=None):
                        j.ll.ril[o]+=1
    j.ril=[0,0,0]
    j.lil=[0,0,0]
    j.uil=[0,0,0]
    j.dil=[0,0,0]
    
```

```python
def con(j1,j2,j3,j4):
    print("       |     |         |     |")
    print("       |{} {}|         |{} {}|".format("  ",str(j1.uil[0])+"a","  ",str(j4.uil[0])+"a"))
    print("       |{} {}|         |{} {}|".format("  ",str(j1.uil[1])+"b","  ",str(j4.uil[1])+"b"))
    print("       |{} {}|         |{} {}|".format("  ",str(j1.uil[2])+"c","  ",str(j4.uil[2])+"c"))
    print("-------       ---------       ---------")
    print(" {}{}{}        {} {} {}       {}{}{}".format(str(j1.lil[0])+"a",str(j1.lil[1])+"b",str(j1.lil[2])+"c",str(j4.lil[0])+"a",str(j4.lil[1])+"b",str(j4.lil[2])+"c","  ","  ","  "))
    print(" {}{}{}        {} {} {}       {}{}{}".format("  ","  ","  ",str(j1.ril[2])+"c",str(j1.ril[1])+"b",str(j1.ril[0])+"a",str(j4.ril[0])+"a",str(j4.ril[1])+"b",str(j4.ril[2])+"c"))
    print("-------       ---------       ---------")
    print("       |     |         |     |")
    print("       |{} {}|         |{} {}|".format(str(j1.dil[2])+"c",str(j2.uil[0])+"a",str(j4.dil[2])+"c",str(j3.uil[0])+"a"))
    print("       |{} {}|         |{} {}|".format(str(j1.dil[1])+"b",str(j2.uil[1])+"b",str(j4.dil[1])+"b",str(j3.uil[1])+"b"))
    print("       |{} {}|         |{} {}|".format(str(j1.dil[0])+"a",str(j2.uil[2])+"c",str(j4.dil[0])+"a",str(j3.uil[2])+"c"))
    print("-------       ---------       ---------")
    print(" {}{}{}        {} {} {}       {}{}{}".format(str(j2.lil[0])+"a",str(j2.lil[1])+"b",str(j2.lil[2])+"c",str(j3.lil[0])+"a",str(j3.lil[1])+"b",str(j3.lil[2])+"c","  ","  ","  "))
    print(" {}{}{}        {} {} {}       {}{}{}".format("  ","  ","  ",str(j2.ril[2])+"c",str(j2.ril[1])+"b",str(j2.ril[0])+"a",str(j3.ril[0])+"a",str(j3.ril[1])+"b",str(j3.ril[2])+"c"))
    print("-------       ---------       ---------")
    print("       |{} {}|         |{} {}|".format(str(j2.dil[2])+"c","  ",str(j3.dil[2])+"c","  "))
    print("       |{} {}|         |{} {}|".format(str(j2.dil[1])+"b","  ",str(j3.dil[1])+"b","  "))
    print("       |{} {}|         |{} {}|".format(str(j2.dil[0])+"a","  ",str(j3.dil[0])+"a","  "))
    print("       |     |         |     |")
```

```python
def plot_vehicle_distribution(junctions):
    labels = ['Junction 1', 'Junction 2', 'Junction 3', 'Junction 4']
    lil = [j.lil[0] + j.lil[1] + j.lil[2] for j in junctions]
    ril = [j.ril[0] + j.ril[1] + j.ril[2] for j in junctions]
    uil = [j.uil[0] + j.uil[1] + j.uil[2] for j in junctions]
    dil = [j.dil[0] + j.dil[1] + j.dil[2] for j in junctions]
    
    x = range(len(labels))
    plt.bar(x, lil, width=0.2, label='Left Inflow', align='center')
    plt.bar([p + 0.2 for p in x], ril, width=0.2, label='Right Inflow', align='center')
    plt.bar([p + 0.4 for p in x], uil, width=0.2, label='Up Inflow', align='center')
    plt.bar([p + 0.6 for p in x], dil, width=0.2, label='Down Inflow', align='center')
    
    plt.xlabel('Junctions')
    plt.ylabel('Number of Vehicles')
    plt.title('Vehicle Distribution at Junctions')
    plt.xticks([p + 0.3 for p in x], labels)
    plt.legend()
    plt.show()
```

```python
def plot_junctions(junctions):
    G = nx.DiGraph()
    
    G.add_node('Junction 1')
    G.add_node('Junction 2')
    G.add_node('Junction 3')
    G.add_node('Junction 4')
    
    G.add_edge('Junction 1', 'Junction 2')
    G.add_edge('Junction 1', 'Junction 3')
    G.add_edge('Junction 2', 'Junction 1')
    G.add_edge('Junction 2', 'Junction 4')
    G.add_edge('Junction 3', 'Junction 1')
    G.add_edge('Junction 3', 'Junction 4')
    G.add_edge('Junction 4', 'Junction 3')
    G.add_edge('Junction 4', 'Junction 2')
    
    pos = nx.spring_layout(G)
    nx.draw(G, pos, with_labels=True, node_size=2000, node_color='skyblue', font_size=12, font_weight='bold')
    plt.show()
```

```python
def jpf():
    zc=0
    jpl=[]
    for i in jl:
        si=prioj(i)
        jpl.append(si)
        if(si==0):
            zc+=1
    if(zc==4):
        return "end"
    m=0
    ind=0
    for i in range(4):
        if(jpl[i]>m):
            m=jpl[i]
            ind=i
    con(j1,j2,j3,j4)
    return ind
```

```python
while(True):
    v=jpf()
    print("junction pro",v)
    if(v=="end"):
        print("end")
        break
    else:
        tc(jl[v])
        plot_vehicle_distribution(jl)
        plot_junctions(jl)
```
